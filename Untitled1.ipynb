```python
import pandas as pd
from sklearn.datasets import load_iris

iris=load_iris()

data=pd.DataFrame(iris.data,columns=iris.feature_names)
data['Label']=iris.target
data.head(150)

data['Label']=data['Label'].apply(lambda x:'Yes' if x==0 else 'No')

def find_s_algorithm(data,target_column):
    attributes=data.drop(columns=[target_column])
    target=data[target_column]
    hypothesis=['0']*attributes.shape[1]
    for i in range(len(data)):
        if target.iloc[i]=='Yes':
            for j in range(len(hypothesis)):
                if hypothesis[j]=='0':
                    hypothesis[j]=attributes.iloc[i,j]
                elif hypothesis[j]!=attributes.iloc[i,j]:
                    hypothesis[j]='?'
    return hypothesis
final_hypothesis=find_s_algorithm(data,'Label')
print("Final Hypothesis")
print(final_hypothesis)
```
